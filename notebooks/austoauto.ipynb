```python
print("Dataset Shape:", df.shape)

print("\nColumn Names:")
print(df.columns.tolist())

print("\nData Types:")
print(df.dtypes)
```

```python
print("Missing Values per Column:\n")
print(df.isnull().sum())
```

```python
print("Numerical Summary:\n")
print(df.describe())
```

```python
print("\nFull Summary (including categorical):\n")
print(df.describe(include='all'))
```

```python
duplicate_count = df.duplicated().sum()
print("Number of duplicate rows:", duplicate_count)
```

```python
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    print(f"\n{col} - Unique values:")
    print(df[col].unique())
```

```python
df['Gender'] = df['Gender'].replace({'Femal': 'Female', 'Femle': 'Female'})
```

```python
num_cols = ['Age', 'No_of_Dependents', 'Salary', 'Partner_salary', 'Total_salary', 'Price']

for col in num_cols:
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
cat_cols = ['Gender', 'Profession', 'Marital_status', 'Education',
            'Personal_loan', 'House_loan', 'Partner_working', 'Make']

for col in cat_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index, palette='pastel')
    plt.title(f'Count of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
num_cols = ['Age', 'No_of_Dependents', 'Salary', 'Partner_salary', 'Total_salary', 'Price']

corr_matrix = df[num_cols].corr()

plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Between Numerical Variables')
plt.show()
```

```python
pairs = [
    ('Gender', 'Price'),
    ('Profession', 'Price'),
    ('Make', 'Price'),
    ('Education', 'Salary'),
    ('Marital_status', 'Total_salary')
]

for cat, num in pairs:
    plt.figure(figsize=(7, 4))
    sns.boxplot(data=df, x=cat, y=num, palette='Set2')
    plt.title(f'{num} by {cat}')
    plt.xlabel(cat)
    plt.ylabel(num)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
cat_pairs = [
    ('Gender', 'Personal_loan'),
    ('Gender', 'House_loan'),
    ('Marital_status', 'Partner_working'),
    ('Profession', 'House_loan'),
    ('Education', 'Personal_loan')
]

for x, hue in cat_pairs:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=x, hue=hue, palette='pastel')
    plt.title(f'{hue} by {x}')
    plt.xlabel(x)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Gender', hue='Make', palette='Set2')
plt.title('Car Make Preference by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Car Make')
plt.tight_layout()
plt.show()
```

```python
total_salaried = df[df['Profession'] == 'Salaried'].shape[0]

salaried_sedan = df[(df['Profession'] == 'Salaried') & (df['Make'] == 'Sedan')].shape[0]

likelihood = (salaried_sedan / total_salaried) * 100
print(f"Likelihood of a salaried person buying a Sedan: {likelihood:.2f}%")
```

```python
salaried_males = df[(df['Gender'] == 'Male') & (df['Profession'] == 'Salaried')]

suv_count = salaried_males[salaried_males['Make'] == 'SUV'].shape[0]
sedan_count = salaried_males[salaried_males['Make'] == 'Sedan'].shape[0]

print(f"SUV purchases by salaried males: {suv_count}")
print(f"Sedan purchases by salaried males: {sedan_count}")
```

```python
gender_price_stats = df.groupby('Gender')['Price'].agg(['mean', 'median', 'count'])
print(gender_price_stats)
```

```python
total_spent_loan = df[df['Personal_loan'] == 'Yes']['Price'].sum()

print(f"Total money spent by personal loan holders: ₹{total_spent_loan:,.2f}")
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x='Partner_working', y='Price', palette='Set2')
plt.title('Price Distribution by Partner Working Status')
plt.xlabel('Partner Working')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

```python
partner_price_stats = df.groupby('Partner_working')['Price'].agg(['mean', 'median', 'count'])
print(partner_price_stats)
```
